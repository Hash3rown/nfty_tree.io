```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["decentraland", "axie infinity", "chaingames", "gods unchained"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 12-m', geo='', gprop='')
mix_12m = pytrends.interest_over_time()
```

```python
mix_12m.plot(title = "Google search trend (12 months)")
```
