______________________________________

** Create a total column for each row **

______________________________________

```python
total_daily_market_cap = combined_market_cap_df.sum(axis=1)
total_daily_market_cap.tail()
```

```python
market_cap_daily_proportions = combined_market_cap_df.divide(total_daily_market_cap, axis=0)
market_cap_daily_proportions.head()
```

```python
# test to make sure everything adds up perfectly to one
one_test = market_cap_daily_proportions.sum(axis=1)
one_test.tail()
```
